```python
# Importa a biblioteca pandas para trabalhar com dados em formato de tabela.
import pandas as pd
```

```python
# Carrega os dados de um arquivo CSV (salaries.csv) direto do GitHub para uma tabela chamada 'df'.
df = pd.read_csv("https://raw.githubusercontent.com/isaacspj/pyton/refs/heads/main/pyton.csv")
```

```python
# Mostra as primeiras 5 linhas da tabela 'df' para ver como os dados são.
df.head()
```

```python
# Mostra um resumo dos dados, como quantas linhas, colunas e o tipo de informação em cada coluna.
df.info()
```

```python
# Mostra o tamanho da tabela, ou seja, quantas linhas e colunas ela tem.
df.shape
```

```python
# Pega o número de linhas e colunas da tabela 'df'.
linhas,colunas = df.shape[0], df.shape[1]
# Mostra o número de linhas.
print("linhas= ",linhas)
# Mostra o número de colunas.
print("colunas= ",colunas)
```

```python
# Mostra o nome de todas as colunas da tabela 'df'.
df.columns
```

```python
# Conta quantos de cada nível de experiência existem na coluna 'experience_level'.
df["experience_level"].value_counts()
```

```python
# Conta quantos de cada tamanho de empresa existem na coluna 'company_size'.
df["company_size"].value_counts()
```

```python
# Cria um 'dicionário' para traduzir as letras dos tamanhos de empresa (M, L, S) para palavras (Médio, Grande, Pequeno).
mapa_tamanho_empresa = {
    'M': 'Medio',
    'L': 'Grande',
    'S': 'Pequeno'
}
# Troca o nome da coluna 'company_size' para 'tamanho_empresa'.
df = df.rename(columns={'company_size': 'tamanho_empresa'})
# Aplica a tradução aos valores da coluna 'tamanho_empresa'.
df['tamanho_empresa'] = df['tamanho_empresa'].map(mapa_tamanho_empresa)
# Mostra quantos de cada tamanho de empresa existem depois da tradução.
display(df['tamanho_empresa'].value_counts())
```

```python
# Lista os nomes das colunas em inglês e seus equivalentes em português.
colunas_traduzidas = {
    'work_year': 'ano_trabalho',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_funcionario',
    'remote_ratio': 'percentual_remoto',
    'company_location': 'localizacao_empresa'
    # 'company_size': 'tamanho_empresa' # Essa coluna já foi renomeada antes.
}
# Renomeia as colunas da tabela 'df' usando a lista de traduções.
df = df.rename(columns=colunas_traduzidas)
# Mostra os novos nomes das colunas para confirmar que foram traduzidos.
print(df.columns)
```

```python
# Mostra as primeiras 5 linhas da tabela 'df' depois de renomear as colunas e traduzir alguns valores.
df.head()
```

```python
# Dicionário para traduzir as siglas de tipo de emprego (FT, PT, CT, FL) para português.
mapa_tipo_emprego = {
    'FT': 'Tempo Integral',
    'PT': 'Meio Período',
    'CT': 'Contrato',
    'FL': 'Freelancer'
}

# Mostra quantos de cada tipo de emprego existem antes da tradução.
print("Contagem antes da tradução:")
display(df['tipo_emprego'].value_counts())

# Traduz os valores da coluna 'tipo_emprego' usando o dicionário.
df['tipo_emprego'] = df['tipo_emprego'].map(mapa_tipo_emprego)

# Mostra quantos de cada tipo de emprego existem depois da tradução.
print("\nContagem depois da tradução:")
display(df['tipo_emprego'].value_counts())
```

```python
# Dicionário para traduzir as siglas de nível de experiência (SE, MI, EN, EX) para português.
mapa_nivel_experiencia = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

# Mostra quantos de cada nível de experiência existem antes da tradução.
print("Contagem antes da tradução:")
display(df['nivel_experiencia'].value_counts())

# Traduz os valores da coluna 'nivel_experiencia' usando o dicionário.
df['nivel_experiencia'] = df['nivel_experiencia'].map(mapa_nivel_experiencia)

# Mostra quantos de cada nível de experiência existem depois da tradução.
print("\nContagem depois da tradução:")
display(df['nivel_experiencia'].value_counts())
```
